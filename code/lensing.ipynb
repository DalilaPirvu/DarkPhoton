```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
myimage = mpimg.imread('images/CookieMonster.jpg')
print(np.shape(myimage))
myimage = myimage[:,:,1]
```

```python
def f(x, y):
    x_s = -x
    y_s = np.sign(y)*(np.abs(y))**0.6
    return x_s, y_s
```

```python
# create mesh for image
x, y = np.meshgrid(np.linspace(-10,10,np.shape(myimage)[0]),np.linspace(-10,10,np.shape(myimage)[1]),sparse=True)
# apply coordinate transformation
x_lens, y_lens = f(x[0,:], y[:,0])
# interpolate image onto transformed coordinates
g = interpolate.interp2d(x_lens, y_lens, myimage)
# evaluate function on the mesh
lensimage = g(x[0,:], y[:,0])
```

```python
# visualize coordinate transformation

plt.figure()
plt.plot(x[0,:], label='Original x')
plt.plot(y[:,0], label='Original y')
plt.plot(x_lens,'--', label='Lensed x')
plt.plot(y_lens,'-.', label='Lensed y')
plt.legend()
plt.show()
```

```python
# visualize original and lensed pictures

plt.figure()
print(np.shape(myimage))
plt.imshow(myimage)

plt.figure()
plt.imshow(lensimage)
plt.show()
```
